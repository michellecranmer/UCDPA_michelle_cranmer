```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#life Expectancy vs Alcohol Intake.
```

```python
life_exp=pd.read_csv('lifeexpectancy-verbose.csv')
```

```python
def cleaning():
    life_exp.fillna('unknown')
```

```python
life_exp.info()
```

```python
life_exp.head()
```

```python
max_life=life_exp.groupby(['CountryDisplay'])['DisplayValue'].max()
```

```python
max_life_true = max_life.reset_index(drop=False)
max_life_true.head()
```

```python
alcohol=pd.read_csv('drinks.csv')
```

```python
cleaning()
```

```python
alcohol.head(4)
```

```python
alcohol.rename(columns = {'country':'CountryDisplay'}, inplace = True) 
```

```python
alco_df = alcohol.reset_index(drop=False)
```

```python
alco_df.shape
```

```python
alco_life_exp=pd.merge(max_life_true, alco_df, how='outer', on=['CountryDisplay'])
```

```python
alco_life_exp.head()
```

```python
del alco_life_exp["index"]
```

```python
top_10=alco_life_exp.nlargest(10,'total_litres_of_pure_alcohol')
```

```python
top_10.head()
```

```python
fig,ax = plt.subplots()
plt.rcParams["figure.figsize"] = (25,15)
plt.rcParams['font.size'] = '20'
ax.plot(top_10['CountryDisplay'],top_10['total_litres_of_pure_alcohol'],marker='^',markersize=20,linestyle='-',color='red',label='Alcohol',linewidth=4)
ax2=ax.twinx()
ax2.plot(top_10['CountryDisplay'],top_10['DisplayValue'],marker='o',linestyle=':',color='darkblue',label='Life Expectancy',linewidth=4,markersize=20)
ax.set_xlabel('Country')
ax.set_ylabel('Total Pure Alcohol (litres)')
ax2.set_ylabel('Life Expectancy')
plt.title('Life Expectancy vs Alcohol Intake')
plt.legend()
plt.show()
```

```python
beer_fan=alco_life_exp.nlargest(10,'beer_servings')
```

```python
wine_fan=alco_life_exp.nlargest(10,'wine_servings')
```

```python
spirits_fan=alco_life_exp.nlargest(10,'spirit_servings')
```

```python
fig,ax=plt.subplots(2,2)
plt.rcParams["figure.figsize"] = (25,15)
plt.rcParams['font.size'] = '15'
ax[0,0].plot(top_10['CountryDisplay'],top_10['total_litres_of_pure_alcohol'],marker='o',linestyle='--',color='r')
ax[0,1].plot(beer_fan['CountryDisplay'],beer_fan['beer_servings'],marker='v',linestyle='--',color='b')
ax[1,0].plot(wine_fan['CountryDisplay'],wine_fan['wine_servings'],marker='s',linestyle='--',color='g')
ax[1,1].plot(spirits_fan['CountryDisplay'],wine_fan['wine_servings'],marker='D',linestyle='--',color='y')
ax[0,0].set_xlabel('Country')
ax[0,1].set_xlabel('Country')
ax[1,0].set_xlabel('Country')
ax[1,1].set_xlabel('Country')
ax[0,0].set_ylabel('Total Pure Alcohol (litres)')
ax[0,1].set_ylabel('beer_servings')
ax[1,0].set_ylabel('wine_servings')
ax[1,1].set_ylabel('spirit_servings')
ax[0,0].tick_params(axis='x', rotation=60)
ax[0,1].tick_params(axis='x', rotation=60)
ax[1,0].tick_params(axis='x', rotation=60)
ax[1,1].tick_params(axis='x', rotation=60)
plt.subplots_adjust(hspace=0.6)
plt.show()
```

```python
# API Of Annual GPD In The USA (A 20 Year Review)
```

```python
REAL_GDP = requests.get('https://www.alphavantage.co/query?function=REAL_GDP&interval=annual&apikey=VL3RK94YY336VZ0Y')
```

```python
REAL_GDP.status_code
```

```python
data_gdp=REAL_GDP.json()
```

```python
print(data_gdp)
```

```python
date_value=(data_gdp['data'])
```

```python
dataframe = pd.DataFrame(date_value,columns=['date', 'value'])
```

```python
dataframe["date"] = dataframe["date"].astype("datetime64")
```

```python
dataframe = dataframe.set_index("date")
```

```python
dataframe.head()
```

```python
last_20_years=dataframe.iloc[:20]
```

```python
fig,ax=plt.subplots()
plt.plot(last_20_years["value"], marker='o')
plt.xlabel("date")
plt.ylabel('value')
plt.gca().invert_yaxis()
plt.title("Annual Real Gross Domestic Product (bn)")
plt.show()
```
